```

from pathlib import Path
assert Path("../artifacts/keras_model.keras").exists()
assert Path("../artifacts/scaler.json").exists()
```

```

import subprocess
Path("../web/model").mkdir(parents=True, exist_ok=True)
subprocess.run([
    "tensorflowjs_converter",
    "--input_format=keras",
    "../artifacts/keras_model.keras",
    "../web/model"
], check=True)
```

```

import shutil
shutil.copy("../artifacts/scaler.json", "../web/scaler.json")
print("TF.js model ready")
```
