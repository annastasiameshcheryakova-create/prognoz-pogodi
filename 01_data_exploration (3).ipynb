```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

DATA_PATH = "../data/weather_kriviyrih.csv"
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
df.info()
```

```python
df['time'] = pd.to_datetime(df['time'], errors='coerce')
df = df.dropna(subset=['time']).sort_values('time').reset_index(drop=True)
print('Duplicates:', df.duplicated(subset=['time']).sum())
num_cols = [c for c in df.columns if c!='time']
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
df[num_cols] = df[num_cols].interpolate(limit_direction='both').fillna(method='bfill').fillna(method='ffill')
display(df.isna().mean().head())
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['time'], df['temperature_2m'])
plt.title('Temperature trend (2025)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
